```python
import numpy as np
from scipy import stats

# Dataset with the outlier (50)
data_outlier = np.array([92, 85, 50, 92, 88])
n = len(data_outlier)

print(f"Dataset with Outlier: {data_outlier}\n")

### Central Tendency ###
mean_outlier = np.mean(data_outlier)
median_outlier = np.median(data_outlier)
mode_outlier = stats.mode(data_outlier, keepdims=False).mode

print(f"1. Mean: {mean_outlier:.2f}")
print(f"2. Median: {median_outlier:.2f}")
print(f"3. Mode: {mode_outlier}")

print("-" * 30)

### Dispersion ###
range_outlier = np.max(data_outlier) - np.min(data_outlier)
sample_variance_outlier = np.var(data_outlier, ddof=1)
sample_std_dev_outlier = np.std(data_outlier, ddof=1)
mad_outlier = np.mean(np.abs(data_outlier - mean_outlier))

print(f"1. Range: {range_outlier}")
print(f"2. Sample Variance (s^2): {sample_variance_outlier:.2f}")
print(f"3. Sample Standard Deviation (s): {sample_std_dev_outlier:.3f}")
print(f"4. Mean Absolute Deviation (MAD): {mad_outlier:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# --- 1. Datasets ---
data_original = np.array([92, 85, 78, 92, 88])
data_outlier = np.array([92, 85, 50, 92, 88])

# --- 2. Calculate Statistics (for plotting) ---
# Original Data
mean_orig = np.mean(data_original)
median_orig = np.median(data_original)
std_orig = np.std(data_original, ddof=1)

# Outlier Data
mean_outlier = np.mean(data_outlier)
median_outlier = np.median(data_outlier)
std_outlier = np.std(data_outlier, ddof=1)

print(f"Original Data Mean: {mean_orig:.2f}, Median: {median_orig:.2f}, Std Dev: {std_orig:.2f}")
print(f"Outlier Data Mean: {mean_outlier:.2f}, Median: {median_outlier:.2f}, Std Dev: {std_outlier:.2f}\n")

# --- 3. Visualization ---
plt.style.use('seaborn-v0_8-darkgrid') # A nice aesthetic style for plots

fig, axes = plt.subplots(3, 2, figsize=(14, 15), sharex=True) # 3 rows, 2 columns of subplots
fig.suptitle('Impact of Outlier on Statistics and Visualizations', fontsize=18, y=1.02)


### Row 1: Histograms with Mean & Median ###
# Original Data Histogram
sns.histplot(data_original, kde=True, ax=axes[0, 0], color='skyblue', bins=5)
axes[0, 0].axvline(mean_orig, color='red', linestyle='--', label=f'Mean: {mean_orig:.2f}')
axes[0, 0].axvline(median_orig, color='green', linestyle='-', label=f'Median: {median_orig:.2f}')
axes[0, 0].set_title('Original Data Distribution')
axes[0, 0].set_xlabel('Scores')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].legend()

# Outlier Data Histogram
sns.histplot(data_outlier, kde=True, ax=axes[0, 1], color='lightcoral', bins=5)
axes[0, 1].axvline(mean_outlier, color='red', linestyle='--', label=f'Mean: {mean_outlier:.2f}')
axes[0, 1].axvline(median_outlier, color='green', linestyle='-', label=f'Median: {median_outlier:.2f}')
axes[0, 1].set_title('Outlier Data Distribution')
axes[0, 1].set_xlabel('Scores')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].legend()


### Row 2: Box Plots ###
# Box plot for Original Data
sns.boxplot(y=data_original, ax=axes[1, 0], color='skyblue', width=0.3)
axes[1, 0].axhline(mean_orig, color='red', linestyle='--', label=f'Mean: {mean_orig:.2f}')
axes[1, 0].axhline(median_orig, color='green', linestyle='-', label=f'Median: {median_orig:.2f}')
axes[1, 0].set_title('Original Data Box Plot')
axes[1, 0].set_ylabel('Scores')
axes[1, 0].legend()

# Box plot for Outlier Data
sns.boxplot(y=data_outlier, ax=axes[1, 1], color='lightcoral', width=0.3)
axes[1, 1].axhline(mean_outlier, color='red', linestyle='--', label=f'Mean: {mean_outlier:.2f}')
axes[1, 1].axhline(median_outlier, color='green', linestyle='-', label=f'Median: {median_outlier:.2f}')
axes[1, 1].set_title('Outlier Data Box Plot')
axes[1, 1].set_ylabel('Scores')
axes[1, 1].legend()


### Row 3: Strip Plots with Standard Deviation indication ###
# Strip Plot for Original Data
sns.stripplot(x=['Original'] * len(data_original), y=data_original, ax=axes[2, 0], color='skyblue', size=10, jitter=0.1)
axes[2, 0].axhline(mean_orig, color='red', linestyle='--', label=f'Mean: {mean_orig:.2f}')
axes[2, 0].axhspan(mean_orig - std_orig, mean_orig + std_orig, color='red', alpha=0.1, label=f'Mean $\pm$ Std Dev ({std_orig:.2f})')
axes[2, 0].axhline(median_orig, color='green', linestyle='-', label=f'Median: {median_orig:.2f}')
axes[2, 0].set_title('Original Data Points')
axes[2, 0].set_ylabel('Scores')
axes[2, 0].legend()

# Strip Plot for Outlier Data
sns.stripplot(x=['Outlier'] * len(data_outlier), y=data_outlier, ax=axes[2, 1], color='lightcoral', size=10, jitter=0.1)
axes[2, 1].axhline(mean_outlier, color='red', linestyle='--', label=f'Mean: {mean_outlier:.2f}')
axes[2, 1].axhspan(mean_outlier - std_outlier, mean_outlier + std_outlier, color='red', alpha=0.1, label=f'Mean $\pm$ Std Dev ({std_outlier:.2f})')
axes[2, 1].axhline(median_outlier, color='green', linestyle='-', label=f'Median: {median_outlier:.2f}')
axes[2, 1].set_title('Outlier Data Points')
axes[2, 1].set_ylabel('Scores')
axes[2, 1].legend()


plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Adjust layout to prevent title overlap
plt.show()
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
sns.set_style("whitegrid")

# Create a sample Swiggy dataset
data = {
    'OrderID': np.arange(1001, 1031),
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Delhi'] * 6,
    'Cuisine': ['North Indian', 'Fast Food', 'South Indian', 'Continental', 'Beverages'] * 6,
    'Avg_Meal_Price_INR': np.random.randint(150, 600, 30),
    'Customer_Rating': np.round(np.random.uniform(3.0, 5.0, 30), 1),
    'Preparation_Time_Min': np.random.randint(15, 45, 30),
    'Rider_Distance_KM': np.round(np.random.uniform(1.5, 8.0, 30), 1),
}
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Calculate Total Delivery Time (Preparation + a function of Distance)
# Simple model: Total Time = Prep_Time + (Distance * 4 min/km) + random noise
df['Total_Delivery_Time_Min'] = df['Preparation_Time_Min'] + (df['Rider_Distance_KM'] * 4) + np.random.randint(1, 5, 30)
df['Is_Late'] = (df['Total_Delivery_Time_Min'] > 45).astype(int) # Flag if delivery exceeded 45 mins
```

```python
print("--- Sample Swiggy Data Snapshot (First 5 Rows) ---")
print(df.head())
print("\n" + "="*80 + "\n")
```

```python
df.info()
```

```python
print(df[['Preparation_Time_Min', 'Rider_Distance_KM',
          'Total_Delivery_Time_Min', 'Is_Late']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization
stats_summary = df[['Avg_Meal_Price_INR', 'Customer_Rating', 'Total_Delivery_Time_Min']].agg(['mean', 'median', 'std', 'min', 'max'])

print("--- Descriptive Statistics Summary ---")
print(stats_summary)
print("\n" + "="*80 + "\n")

# Visualization: Distribution of Delivery Time
plt.figure(figsize=(8, 5))
sns.histplot(df['Total_Delivery_Time_Min'], kde=True, bins=10, color='skyblue')
plt.title('Distribution of Total Delivery Time (Minutes)')
plt.xlabel('Total Delivery Time (Min)')
plt.ylabel('Frequency (Number of Orders)')
plt.show()
```

```python
# Calculate the probability of a late order
late_probability = df['Is_Late'].mean()

# Group by city to find late probability per city
city_late_prob = df.groupby('City')['Is_Late'].mean().sort_values(ascending=False)

print(f"--- Probability of an Order Being Late (> 45 min) ---")
print(f"P(Late) = {late_probability:.2f} (i.e., {late_probability*100:.0f}% of orders are late)")
print("\n--- Late Order Probability by City ---")
print(city_late_prob)
print("\n" + "="*80 + "\n")

# Visualization: Bar plot for Late Probability by City
plt.figure(figsize=(7, 5))
sns.barplot(x=city_late_prob.index, y=city_late_prob.values, palette='Reds_d')
plt.title('Probability of Late Order by City')
plt.ylabel('P(Late Order)')
plt.show()

#
```

```python
from scipy import stats

# Perform Linear Regression to find the correlation
slope, intercept, r_value, p_value, std_err = stats.linregress(
    df['Total_Delivery_Time_Min'],
    df['Customer_Rating']
)

# R-value measures the correlation: closer to -1 means strong negative correlation
print(f"--- Correlation between Delivery Time and Customer Rating ---")
print(f"Pearson Correlation (R-value): {r_value:.2f}")

# Visualization: Scatter Plot with Regression Line
plt.figure(figsize=(8, 5))
sns.regplot(x='Total_Delivery_Time_Min', y='Customer_Rating', data=df,
            line_kws={"color": "red"}, scatter_kws={'alpha':0.6})
plt.title(f'Delivery Time vs. Customer Rating (R = {r_value:.2f})')
plt.xlabel('Total Delivery Time (Min)')
plt.ylabel('Customer Rating')
plt.show()

#
```

```python
# Calculate the probability of the Late Delivery
total_deliveries = len(df)
late_deliveries = df["is_late"].sum()
probability_late = late_deliveries / total_deliveries
probability_on_time = 1 - probability_late

print(f"Total Deliveries in Sample: {total_deliveries}")
print(f"Number of Late Deliveries: {late_deliveries}")
print(f"1. Probability of a delivery being Late: {probability_late:.2f} ({probability_late*100:.0f}%)")
print(f"2. Probability of a delivery being On Time: {probability_on_time:.2f} ({probability_on_time*100:.0f}%)\n")

print("\nInterpretation:")
print(f"There is a {probability_late*100:.0f}% chance that any given order will be late. This high number suggests a need to improve logistics or adjust estimated delivery times.\n")

# Visualization for Simple Probability: Pie Chart
plt.figure(figsize=(6, 6))
labels = ['Late Delivery', 'On Time']
sizes = [probability_late, probability_on_time]
colors = ['tomato', 'mediumseagreen']
explode = (0.1, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.0f%%', shadow=True, startangle=140)
plt.title('Delivery Performance Breakdown')
plt.show()
print("Visualization 5: Pie Chart showing the percentage of late versus on-time deliveries.")
```
